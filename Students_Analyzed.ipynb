```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

increaser_female = 0
increaser_male = 0
Data_s = pd.read_csv("StudentsPerformance.csv")

female = Data_s.loc[Data_s["gender"] == "female"].count()[0]
male = Data_s.loc[Data_s["gender"] == "male"].count()[0]


for row, column in Data_s.iterrows():
  gender = column["gender"]
  score = column["writing score"]
  if gender == "female":
    increaser_female += score
  else:
    increaser_male += score

average_female = increaser_female / female
average_male = increaser_male / male
average_marks = [average_female, average_male]

x_axis = ["Female", "Male"]

plt.bar(x_axis, average_marks,)
plt.title("Average Writing Score by Gender", fontdict={"fontsize": 20})
plt.xlabel("Gender")
plt.ylabel("Average Writing Score")

for i in range(len(x_axis)):
    plt.text(i, average_marks[i] + 1, f"{average_marks[i]:.1f}", ha='center')
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

no_course_sum = 0
no_courser = 0
courser_sum = 0
courser = 0
Data_s = pd.read_csv("StudentsPerformance.csv")

for row, column in Data_s.iterrows():
  which = column["test preparation course"]
  score_average = (column["math score"] + column["reading score"] + column["writing score"]) / 3
  if which == "completed":
    courser_sum += score_average
    courser += 1
  else:
    no_course_sum += score_average
    no_courser += 1

courser_average = courser_sum / courser
no_courser_average = no_course_sum / no_courser
results = [courser_average, no_courser_average]
x_axis = ["Completed", "Not Completed"]
plt.bar(x_axis, results)
plt.title("Average Score by Course", fontdict={"fontsize": 20})

plt.xlabel("Course", fontdict={"fontsize": 8, "style": "italic"})
plt.ylabel("Average Score")

for i in range(len(x_axis)):
    plt.text(i, results[i] + 1, f"{results[i]:.1f}", ha='center')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

Data_s = pd.read_csv("StudentsPerformance.csv")

group_A = 0
group_B = 0
group_C = 0
#
group_A_sum_all = 0
group_B_sum_all = 0
group_C_sum_all = 0
#
group_A_sum_math = 0
group_B_sum_math = 0
group_C_sum_math = 0
#
group_A_sum_reading = 0
group_B_sum_reading = 0
group_C_sum_reading = 0
#
group_A_sum_writing = 0
group_B_sum_writing = 0
group_C_sum_writing = 0


for row, column in Data_s.iterrows():
  group = column["race/ethnicity"]
  math_score = column["math score"]
  reading_score = column["reading score"]
  writing_score = column["writing score"]
  score_average = (math_score + reading_score + writing_score) / 3
  if group == "group A":
    group_A_sum_all += score_average
    group_A_sum_math += math_score
    group_A_sum_reading += reading_score
    group_A_sum_writing += writing_score
    group_A += 1
  elif group == "group B":
    group_B_sum_all += score_average
    group_B_sum_math += math_score
    group_B_sum_reading += reading_score
    group_B_sum_writing += writing_score
    group_B += 1
  else:
    group_C_sum_all += score_average
    group_C_sum_math += math_score
    group_C_sum_reading += reading_score
    group_C_sum_writing += writing_score
    group_C += 1

group_A_overall = group_A_sum_all / group_A
group_B_overall = group_B_sum_all / group_B
group_C_overall = group_C_sum_all / group_C

group_A_math = group_A_sum_math / group_A
group_B_math = group_B_sum_math / group_B
group_C_math = group_C_sum_math / group_C

group_A_reading = group_A_sum_reading / group_A
group_B_reading = group_B_sum_reading / group_B
group_C_reading = group_C_sum_reading / group_C

group_A_writing = group_A_sum_writing / group_A
group_B_writing = group_B_sum_writing / group_B
group_C_writing = group_C_sum_writing / group_C

g1_y = [group_A_overall, group_B_overall, group_C_overall]
g2_y = [group_A_math, group_B_math, group_C_math]
g3_y = [group_A_reading, group_B_reading, group_C_reading]
g4_y = [group_A_writing, group_B_writing, group_C_writing]

plt.bar(["Group A", "Group B", "Group C"], g1_y)
plt.title("Average Score by Group", fontdict={"fontsize": 20})
plt.xlabel("Groups", fontdict={"fontsize": 8, "style": "italic"})
plt.ylabel("Average Score")
for i in range(len(g1_y)):
    plt.text(i, g1_y[i] + 1, f"{g1_y[i]:.1f}", ha='center')
plt.show()

plt.bar(["Group A", "Group B", "Group C"], g2_y)
plt.title("Average Math Score by Group", fontdict={"fontsize": 20})
plt.xlabel("Groups", fontdict={"fontsize": 8, "style": "italic"})
plt.ylabel("Average Math Score")
for i in range(len(g2_y)):
    plt.text(i, g2_y[i] + 1, f"{g2_y[i]:.1f}", ha='center')
plt.show()

plt.bar(["Group A", "Group B", "Group C"], g3_y)
plt.title("Average Reading Score by Group", fontdict={"fontsize": 20})
plt.xlabel("Groups", fontdict={"fontsize": 8, "style": "italic"})
plt.ylabel("Average Reading Score")
for i in range(len(g3_y)):
    plt.text(i, g3_y[i] + 1, f"{g3_y[i]:.1f}", ha='center')
plt.show()

plt.bar(["Group A", "Group B", "Group C"], g4_y)
plt.title("Average Writing Score by Group", fontdict={"fontsize": 20})
plt.xlabel("Groups", fontdict={"fontsize": 8, "style": "italic"})
plt.ylabel("Average Writing Score")
for i in range(len(g4_y)):
    plt.text(i, g4_y[i] + 1, f"{g4_y[i]:.1f}", ha='center')
plt.show()
```
